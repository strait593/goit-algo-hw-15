### Heaps

Heap is a data structure that can simplify the process of finding largest or smallest elements within a collection. Works according to LIFO (Last In First Out) principle. The heap is a complete binary tree, which means that all levels of the tree are fully filled except possibly for the last level, which is filled from left to right.
