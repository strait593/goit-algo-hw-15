```python
import nltk
import spacy
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from heapq import nlargest

nltk.download("punkt")
nltk.download("stopwords")

nlp = spacy.load("en_core_web_sm")
```

```python
# Open up the text
with open("article.txt", 'r', encoding='utf-8') as article:
    for word in article:
        text = "".join(word)

print(text)
```

```python
# break the text down word by word
doc = nlp(text)
tokens = [token.text for token in doc]

print(tokens)
```

```python
# break the text down sentence by sentence
words = word_tokenize(text)
sentences = sent_tokenize(text)
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
```

```python
# calculate how many times does each word appear within the text
# P.S This is how the model identifies the max likelyhood for the topic of the text
word_freq = {}

for w in words:
    w = w.lower()
    if w not in stop_words and w not in punctuation:
        if w not in word_freq.keys():
            word_freq[w] = 1
        else:
            word_freq[w] += 1

for word in word_freq.keys():
    print(f"{word}: {word_freq[word]}")
```

```python
# score the sentences

sentence_scores = {}

for sentence in sentences:
    for word in word_tokenize(sentence.lower()):
        if word in word_freq.keys():
            if len(sentence.split(' ')) < 30:
                if sentence not in sentence_scores.keys():
                    sentence_scores[sentence] = word_freq[word]
                else:
                    sentence_scores[sentence] += word_freq[word]

for sentence in sentence_scores.keys():
    print(f"{sentence}: {sentence_scores[sentence]}")
```

```python
# now, select top 3 sentences from the entire article
summary_sentences = nlargest(3, sentence_scores, key=sentence_scores.get)

print("Summary:\n {}".format(' '.join(summary_sentences)))
```

### Conclusion

The model does a good job grasping the overall concept of this article about space travel.
The mentions abour space shuttles and the ISS are important to the overall understanding of the article, and the model was able to identify them as such. The model also did a good job of identifying the main points of the article, such as the fact that space travel is expensive and that there are many challenges associated with it.
